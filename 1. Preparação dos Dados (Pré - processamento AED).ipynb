#### 4. Unificação das bases dos diferentes anos
Realizar a união de bases de dados em uma análise multianual é fundamental para garantir uma visão mais completa, consistente e comparável ao longo do tempo. Isso permite unir as informações dos diferentes anos em uma única estrutura, facilitando sua exploração.

```python
gc.collect()

df_unif = pd.concat([df_saeb_2021, df_saeb_2023], ignore_index=True)

df_unif = df_unif.copy()

numericas = [
    "ANO",
    "PUBLICA",
    "LOCALIZACAO",
    "PROFICIENCIA",
    "PROFICIENCIA_SAEB",  
]    
for col in numericas:
    df_unif[col] = pd.to_numeric(
    df_unif[col],
    errors="coerce"
)

categoricas = [
    "SEXO",
    "COR/RACA",
    "ESCOL_MAE",
    "FREQUENCIA_CONVERSA",
    "QTD_COMPUTADOR",
    "QTD_CARRO",
    "GARAGEM",
    "IDADE_INTROD_ESC",     
    "REPROVACAO",
    "TEMPO_TRAB_DOMES",
    "TEMPO_ESTUDO",
    "POS_EF"    
]
for col in categoricas:
    df_unif[col] = df_unif[col].astype("category")

processed_dir = r"C:\Users\emill\Downloads\TCC\processed"

os.makedirs(processed_dir, exist_ok=True)

out_path = os.path.join(processed_dir, "BASE_AED.parquet")

df_unif = pd.DataFrame(df_unif)
df_unif.reset_index(drop=True, inplace=True)

df_unif.to_parquet(out_path, index=False, engine="pyarrow")
```
